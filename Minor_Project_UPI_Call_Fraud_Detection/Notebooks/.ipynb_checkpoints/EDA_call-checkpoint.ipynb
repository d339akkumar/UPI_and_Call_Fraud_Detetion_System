### Import Libraries

```python
# Basic setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# For stats and time handling
from datetime import datetime
import warnings
warnings.filterwarnings('ignore')

# Visualization settings
sns.set(style='whitegrid', palette='Set2')
plt.rcParams['figure.figsize'] = (10,6)
```

### Load Dataset

```python
# Load the dataset
df = pd.read_csv('../dataset/CDR_dataset.csv', parse_dates=['call_start_time'])

# Display basic info
print("Dataset shape:", df.shape)
print("\nColumn names:", df.columns.tolist())

# Preview first few rows
df.head()
```

### Overview 

```python
# Info and data types
df.info()

# Statistical summary
df.describe(include='all').transpose()
```

### Lable Distribution

```python
# Fraud vs Not Fraud count and percentage
label_counts = df['label'].value_counts()
print("Label Distribution:\n", label_counts)
print("\nFraud Percentage:", round(100 * label_counts['Fraud'] / len(df), 2), "%")

# Plot class distribution
sns.countplot(data=df, x='label', palette='Set1')
plt.title("Fraud vs Not Fraud Distribution")
plt.show()
```

### Call Type Analysis

```python
# Call type frequency
call_type_counts = df['call_type'].value_counts()
print(call_type_counts)

# Bar plot
sns.countplot(data=df, x='call_type', hue='label', palette='Set2')
plt.title("Call Type Distribution by Label")
plt.xticks(rotation=45)
plt.show()
```

### Call Duration and Cost Distributions

```python
# Distribution of call duration
sns.histplot(df['call_duration'], bins=50, kde=True)
plt.title("Distribution of Call Duration")
plt.xlabel("Call Duration (seconds)")
plt.show()

# Distribution of call cost
sns.histplot(df['call_cost'], bins=50, kde=True, color='orange')
plt.title("Distribution of Call Cost")
plt.xlabel("Call Cost ()")
plt.show()
```

### Compare Fraud vs Non-Fraud by Numeric Features

```python
# Numeric columns for comparison
num_cols = ['call_duration', 'call_cost', 'cost_per_sec', 
            'distinct_callees_last_24h', 'tower_switch_rate', 
            'repeated_short_calls_last_1h']

for col in num_cols:
    plt.figure(figsize=(8,5))
    sns.boxplot(data=df, x='label', y=col, palette='Set1')
    plt.title(f"{col} by Fraud Label")
    plt.show()
```

### Correlation Analysis

```python
# Convert label to binary for correlation
df_corr = df.copy()
df_corr['label_bin'] = df_corr['label'].map({'Fraud':1, 'Not Fraud':0})

# Compute correlation matrix
corr = df_corr.corr(numeric_only=True)

# Plot heatmap
plt.figure(figsize=(10,8))
sns.heatmap(corr, annot=True, cmap='coolwarm', fmt='.2f')
plt.title("Correlation Heatmap (with Fraud Label)")
plt.show()
```
